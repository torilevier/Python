```python
import pandas as pd
import plotnine as p9
```

```python
surveys = pd.read_csv("data/surveys.csv").dropna()
```

```python
surveys
```

```python
(
    p9.ggplot(
        surveys,
        p9.aes(x="weight", y="hindfoot_length")
        
    )
    + p9.geom_point()
)
            
            
# aes is short for aesthetics
# geom_point means we want to have points on the plot
# actually adding additional things to your plot
```

```python
(
    p9.ggplot(
        surveys,
        p9.aes(x="weight", y="hindfoot_length")
        
    )
    + p9.geom_point(alpha = 0.1)
)

# makes the graph look more transparent
# the default is likely alpha = 1
```

```python
(
    p9.ggplot(
        surveys,
        p9.aes(x="plot_id")
        
    )
    + p9.geom_bar()
)

# plotting a bar graph with only x-axis specified
```

```python
(
    p9.ggplot(
        surveys,
        p9.aes(x="weight", y="hindfoot_length", color="species_id")
        
    )
    + p9.geom_point(alpha = 0.1)
    + p9.xlab("Weight (g)")
    + p9.ylab("Hindfoot length (mm)")
    + p9.scale_x_log10()
)

# divides things up by species and gives them different colors
# the second + is to clean up the x label
# third is to clean up y
# fourth is to make x logarithmic to make it less crowded at the bottom, feel free to comment the line out to see
```

```python
(
    p9.ggplot(
        surveys,
        p9.aes(x="plot_id",fill="sex")
    )
    + p9.geom_bar()
    + p9.scale_fill_manual(["purple", "pink"])
)

# splitting it by sex you can know kind of get this stacked bar chart
# see above for overriding colors to get the ones you prefer
# other colors can be blue or orange or red, etc.
```

```python
(
    p9.ggplot(
        surveys,
        p9.aes(x="species_id", y="weight")
    )
    + p9.geom_boxplot()
)
```

```python
(
    p9.ggplot(
        surveys,
        p9.aes(x="species_id", y="weight")
    )
    + p9.geom_jitter(alpha=0.1)
)

# with geom_point, it is hard to see how much is at each point
# but with geom_jitter, you can get a better sense of the distribution of points
```

```python
(
    p9.ggplot(
        surveys,
        p9.aes(x="weight", y="hindfoot_length", color="species_id")
    )
    + p9.geom_point()
    + p9.facet_wrap("sex")
)

# for facet_wrap you can split it up into two different sexes
```

```python
(
    p9.ggplot(
        surveys,
        p9.aes(x="weight", y="hindfoot_length", color="species_id")
    )
    + p9.geom_point()
    + p9.facet_wrap("plot_id")
)

# this way you can break it up by all the different plot_id's so you can see the mini distributions
```

To find plots you like, look up ggplot galleries online to see what type of plots you can make as well as how to code them!

```python
(
    p9.ggplot(
        surveys,
        p9.aes(x="weight", y="hindfoot_length")
    )
    + p9.geom_point()
    + p9.geom_smooth(method="lm", color="red")
)
```
